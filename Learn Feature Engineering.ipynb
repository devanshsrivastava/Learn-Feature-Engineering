```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1
```

```python
df2 = df1.drop(['area_type','availability','society'],axis = 'columns')
df2
```

```python
df2.shape
```

```python
#To generate a column of price per sqft
def convert_sqft_tonum(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_tonum('2455Perch')
```

```python
df3 = df2.copy()
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_tonum)
```

```python
df3['price_per_sqft'] = df3['price']*100000/df3['total_sqft']
```

```python
df3
```

```python
df3.describe()
```

```python
min_thresh,max_thresh = df3.price_per_sqft.quantile([0.001,0.999])
min_thresh,max_thresh
```

```python
df3[df3.price_per_sqft < min_thresh]
```

```python
df3[df3.price_per_sqft > max_thresh]
```

```python
df4 = df3[(df3.price_per_sqft > min_thresh) & (df3.price_per_sqft < max_thresh)]
```

```python
df4.head()
```
